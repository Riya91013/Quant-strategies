## All indicators functions

```python
def sma_indicator(close, period):
    sma = close.rolling(window=period).mean()
    if close.iloc[-1] > sma.iloc[-1]:
        return "BUY"
    else:
        return "SELL" 
```

```python
def bollinger_bands(close, period, std_dev):
    sma = close.rolling(window=period).mean()
    std = close.rolling(window=period).std()
    upper_band = sma + std_dev * std
    lower_band = sma - std_dev * std
    if close.iloc[-1] < lower_band.iloc[-1]:
        return "BUY"
    elif close.iloc[-1] > upper_band.iloc[-1]:
        return "SELL"
    else:
        return "HOLD"
```

```python
def macd_indicator(close, fast_period, slow_period, signal_period):
    ema_fast = close.ewm(span=fast_period).mean()
    ema_slow = close.ewm(span=slow_period).mean()
    macd = ema_fast - ema_slow
    signal = macd.ewm(span=signal_period).mean()
    if macd.iloc[-1] > signal.iloc[-1]:
        return "BUY"
    else:
        return "SELL"
```

```python
def rsi_indicator(close, period):
    delta = close.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    if rsi.iloc[-1] < 30:
        return "BUY"
    elif rsi.iloc[-1] > 70:
        return "SELL"
    else:
        return "HOLD"
```

```python
def moving_average_cross(close, fast_period, slow_period):
    fast_ma = close.rolling(window=fast_period).mean()
    slow_ma = close.rolling(window=slow_period).mean()
    if fast_ma.iloc[-1] > slow_ma.iloc[-1] and fast_ma.iloc[-2] < slow_ma.iloc[-2]:
        return "BUY"
    elif fast_ma.iloc[-1] < slow_ma.iloc[-1] and fast_ma.iloc[-2] > slow_ma.iloc[-2]:
        return "SELL"
    else:
        return "HOLD"
```

```python
def williams_percent_r(high, low, close):
    highest_high = max(high)
    lowest_low = min(low)
    williams_r = (highest_high - close) / (highest_high - lowest_low) * -100
    if williams_r >= -20:
        return 'BUY'
    else:
        return 'SELL'
```

```python
def fibonacci_retracements(high, low, close):
    highest_high = max(high)
    lowest_low = min(low)
    retracement_levels = [0.0, 0.236, 0.382, 0.5, 0.618, 0.786, 1.0]
    retracements = []
    for level in retracement_levels:
        retracements.append((highest_high - lowest_low) * level + lowest_low)
    if close > retracements[-1]:
        return 'BUY'
    elif close < retracements[0]:
        return 'SELL'
    else:
        for i in range(len(retracements) - 1):
            if retracements[i] < close < retracements[i + 1]:
                return 'SELL'
```

```python
def donchian_channel(high, low):
    highest_high = max(high)
    lowest_low = min(low)
    midpoint = (highest_high + lowest_low) / 2
    if highest_high > midpoint:
        return 'BUY'
    else:
        return 'SELL'
```

```python
def trend_analysis(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    if data.iloc[-1] > mean + std_dev:
        return "Up Trend"
    elif data.iloc[-1] < mean - std_dev:
        return "Down Trend"
    else:
        return "No Trend"
```

## checking the values of all the indicators at 3:25 every day for next 100 days and storing the type returned from each indicator to dataframe and calculating the probability

```python
# axis bank
buy_cnt = 0
sell_cnt = 0
cnt = 0
hold_cnt = 0
for j in range(504416):
    temp = axis_bank['close'][:j]
    high = axis_bank['high'][:j]
    low = axis_bank['low'][:j]
    std_dev = temp.std()
    if axis_bank['time'][j] == "15:25:00":
        
        if sma_indicator(temp,14) == 'BUY':
            buy_cnt += 1
        else:
            sell_cnt += 1
        
        if bollinger_bands(temp,14,std_dev) == 'BUY':
            buy_cnt += 1
        elif bollinger_bands(temp,14,std_dev) == 'SELL':
            sell_cnt += 1
        else:
            hold_cnt += 1
        
        if macd_indicator(temp,12,26,14) == 'BUY':
            buy_cnt += 1
        else:
            sell_cnt += 1
        
        if rsi_indicator(temp,14) == 'BUY':
            buy_cnt += 1
        elif rsi_indicator(temp,14) == 'SELL':
            sell_cnt += 1
        else:
            hold_cnt += 1
            
        if moving_average_cross(temp,12,26) == 'BUY':
            buy_cnt += 1
        elif moving_average_cross(temp,12,26) == 'SELL':
            sell_cnt += 1
        else:
            hold_cnt += 1
            
        if williams_percent_r(high,low,axis_bank['close'][j]) == 'BUY':
            buy_cnt += 1
        else:
            sell_cnt += 1
        
        if fibonacci_retracements(high,low,axis_bank['close'][j]) == 'BUY':
            buy_cnt += 1
        else:
            sell_cnt += 1
            
        if donchian_channel(high,low) == 'BUY':
            buy_cnt += 1
        else:
            sell_cnt += 1
        
        if trend_analysis(temp) == 'Up Trend':
            buy_cnt += 1
        elif trend_analysis(temp) == 'Down Trend':
            sell_cnt += 1
        else:
            hold_cnt += 1
        
        pbuy = buy_cnt /9
        psell = sell_cnt/9
        phold = hold_cnt/9
        axis_bank_res = axis_bank_res.append({'datetime': axis_bank['datetime'][j], 'psell': psell, 'pbuy': pbuy, 'phold': phold, 'sma_type':sma_indicator(temp,14), 'bollinger_type': bollinger_bands(temp,14,std_dev), 'macd_type': macd_indicator(temp,12,26,14) , 'rsi_type': rsi_indicator(temp,14), 'movingavgcross_type': moving_average_cross(temp,12,26), 'william_type': williams_percent_r(high,low,axis_bank['close'][j]), 'fibbonacci_type': fibonacci_retracements(high,low,axis_bank['close'][j]), 'donchian_type': donchian_channel(high,low), 'trend_type': trend_analysis(temp) }, ignore_index = True)
        cnt += 1
        buy_cnt = 0
        sell_cnt = 0
        hold_cnt = 0
        if cnt == 100:
            break
```

```python
def resample_data(df):
    df.set_index(df.datetime, inplace = True)
    df.index = pd.to_datetime(df.index)
    opening=df.resample('5Min',origin= 'start').first()
    opening.drop(['high','low','close','volume'],axis=1,inplace=True)

    high=df.resample('5Min',origin= 'start').max()
    high.drop(['open','low','close','volume','symbol','datetime'],axis=1,inplace=True)

    low=df.resample('5Min',origin= 'start').min()
    low.drop(['open','high','close','volume','symbol','datetime'],axis=1,inplace=True)

    close=df.resample('5Min',origin= 'start').last()
    close.drop(['open','low','high','volume','symbol','datetime'],axis=1,inplace=True)

    df_new = pd.concat([opening,high,low,close], axis = 1)
    df_new.dropna(inplace = True)
    
    return df_new
```

# P(X/Y) = P(X intersection Y)/ P(Y), 
# Here X is the probability that 3:25 open price < 9:16 open price of next day and Y is the probability that buy singal return probability from all the indicators is >= 0.7
